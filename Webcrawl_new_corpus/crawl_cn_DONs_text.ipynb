```python
# import libraries
import sys
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
```

```python
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') # this is must
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chromedriver_autoinstaller.install()
```

```python
# read csv file from and save urls in a list

links_to_DONS = []
with open('/content/drive/MyDrive/Colab Notebooks/Big_data/links_to_DONS.csv', 'r', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for link in reader:
        links_to_DONS.append(link[0])
```

```python
# function to scrape the text in the given DONs links:
def scrape_cn_text(links_to_DONS, driver):
    text = ""
    counter = 0

    for link in links_to_DONS:
        driver.get(link)
        cn_driver.implicitly_wait(5)
        time.sleep(2)

        try:
          elements = driver.find_elements(By.TAG_NAME, "p")
          for element in elements:
            text += element.text

          counter += 1
          print(f"{counter}/{len(links_to_DONS)} done")

        except:
          print(f"error, link {link} failed.")
          counter += 1
          print(f"{counter}/{len(links_to_DONS)} failed")
          pass

    return text
```

```python
# open the driver
cn_driver = webdriver.Chrome(options=chrome_options)
cn_text = scrape_cn_text(links_to_DONS, cn_driver)
cn_driver.quit()
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/Big_data/cn_WHO_text.txt', 'w') as f:
  f.write(cn_text)
```
